```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder , MinMaxScaler , StandardScaler
```

```python
data = pd.read_csv("D:\\titanic_seaborn.csv")
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['age'].fillna(data['age'].median() , inplace = True)
```

```python

data['deck'].fillna('Unknown' , inplace = True)
```

```python
data = data.dropna()
```

```python
data['age'] = data['age'].astype(int)
```

```python
data.info()
```

```python
data
```

```python
m = MinMaxScaler()
data['normalized_age'] = m.fit_transform(data[['age']])
```

```python
data
```

```python
data.drop(columns = ['who' , 'adult_male'] , inplace = True)
```

```python
l = LabelEncoder()
data['sex_encoded'] = l.fit_transform(data['sex'])
```

```python
data = pd.get_dummies(data , columns = ['embark_town'])
```

```python
s = MinMaxScaler()
data['normalized_fare'] = s.fit_transform(data[['fare']])
```
