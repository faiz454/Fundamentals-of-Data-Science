```python
import numpy as np
import pandas as ps
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = ps.read_csv("D:\diabetes.csv")
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns = ['Age'] , inplace = True)
```

```python
data['SkinThickness'].fillna(data['SkinThickness'].median() , inplace = True)
```

```python
data['BMI'].fillna(data['BMI'].mean() , inplace = True)
```

```python
data['BloodPressure'].fillna(data['BloodPressure'].mean() , inplace = True)
```

```python
data['Glucose'].fillna(data['Glucose'].median() , inplace = True)
```

```python
data['DiabetesPedigreeFunction'].fillna(data['DiabetesPedigreeFunction'].median() , inplace = True)
```

```python
data = data.dropna()
```

```python
data
```

```python
scale = MinMaxScaler()
data['SkinThickness'] = scale.fit_transform(data[['SkinThickness']])
```

```python
stand = StandardScaler()
data['BloodPressure'] = scale.fit_transform(data[['BloodPressure']])
```

```python
data['Glucose'] = np.log10(data['Glucose'])
```

```python
data.to_csv('cleaned_data.csv', index=False)
print('Cleaned dataset saved as cleaned_Diabetes_data.csv')
```
